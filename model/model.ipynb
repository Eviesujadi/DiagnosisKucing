```python
from sklearn.naive_bayes import GaussianNB
import pickle
import pandas
```

```python
file = "../data/dataset.csv"
data = pandas.read_csv(file)
data.head()
```

```python
x = data[data.columns[:-1].tolist()]
x
```

```python
y = data["Class"]
y
```

```python
model = GaussianNB()
model.fit(x, y)
```

```python
confidence = model.score(x, y)
confidence
```

```python
pickle.dump(model, open('model.pkl','wb'))
```
